```python
#Autora: Debora Conceição Faria

#Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
#Financiamento Coletivo para Campanhas SP-2020
filename = r'C:\Users\55149\FINANCIAMENTO_COLETIVO_UTF8.csv'
df = pd.read_csv(filename, delimiter=';',
                 error_bad_lines=False)
```

```python
df.shape
```

```python
df.info()
```

```python
df.sample(80)
```

```python
df.groupby('Nome do Doador')['Valor da Doação'].max().sort_values(ascending=False)
```

```python
df.groupby('Nome do Doador')['Valor da Doação'].max().sort_values().tail(10).plot(kind='barh', figsize=(8,8)).set_title('Doações de Candidatos para Campanhas', fontsize=23)
```

```python
df.groupby('Nome da Empresa')['Valor da Doação'].max().sort_values(ascending=False)
```

```python
df.groupby('Nome da Empresa')['Valor da Doação'].max().sort_values().tail(3).plot(kind='barh', figsize=(5,5)).set_title('Empresa Responsável pela Doação', fontsize=20)
```
